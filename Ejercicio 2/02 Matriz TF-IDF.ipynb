```python
import math #para el logaritmo

#corpus
archivos = ['/kaggle/input/corpus1/01_corpus_turismo.txt']

#diccionario 
presencia_por_documento = {}

#limpieza del texto (faltan tildes)
def limpiar_texto(texto):
    signos = ['.', ',', '!', '?', ';', ':', '(', ')', '[', ']', '"', "'"]
    for s in signos:
        texto = texto.replace(s, '') #en singular
    return texto.lower()

#leer corpus
for ruta in archivos:
    with open(ruta, 'r', encoding='utf-8') as f:
        documentos = f.readlines() #division del archivo en documentos

total_documentos = len(documentos)  #calcula el numero de documentos

#enumerar los documentos (id)
for num_documento, documento in enumerate(documentos, 0):
    documento_limpio = limpiar_texto(documento) #quitar simbols
    palabras = documento_limpio.split()
    palabras_unicas = set(palabras) #terminos (palabras sin repetirse)

    #sumar la palabra al diccionario
    for palabra in palabras_unicas:
        if palabra not in presencia_por_documento:
            presencia_por_documento[palabra] = set()
        presencia_por_documento[palabra].add(num_documento)

#mostrar encabezado de la matriz
print(f"{'Término':15s} {'IDF':>6} ", end='')
for i in range(total_documentos):
    print(f"D{i+1:02}", end=' ')
print()

# Mostrar cada término en orden alfabético
for termino in sorted(presencia_por_documento.keys()):
    documentos_donde_aparece = presencia_por_documento[termino]
    #calculo idf
    df = len(documentos_donde_aparece)
    idf = math.log10(total_documentos / df) if df > 0 else 0
    print(f"{termino:15s} {idf:6.4f} ", end='')
    for i in range(total_documentos):
        print(f"{1 if i in documentos_donde_aparece else 0:>3}", end=' ')
    print()

# Imprimir numero total de términos
print(f"\nTotal de términos: {len(presencia_por_documento)}")
```
